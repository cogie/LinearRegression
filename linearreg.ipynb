Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Dataset    X,predictor   Y,target

```python
sal_data = pd.read_csv('C:/Users/Marco/Desktop/MLnew/data/Salary_Data.csv')
X = sal_data.iloc[:, :-1].values
y = sal_data.iloc[:, 1].values
```

Visualize data

```python
sal_data
```

```python
sns.barplot(x = 'YearsExperience', y = 'Salary', data = sal_data)
```

```python
sns.heatmap(sal_data.corr())
```

Splitting the data into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

Fit linear regresion to the training dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

Predict the test set results

```python
y_pred = lr.predict(X_test)
y_pred
```

Visualize train results


```python
plt.scatter(X_train, y_train, color = 'black')
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.title('Salary vs Experience (Train set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

Visualize test results

```python
plt.scatter(X_test, y_test, color = 'black')
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

Finding the Residuals

```python
#Calculate residuals
from sklearn import metrics
print('MAE', metrics.mean_absolute_error(y_test, y_pred))
print('MSE', metrics.mean_squared_error(y_test, y_pred))
print('RMSE', np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
```

RSME value = small == good model
